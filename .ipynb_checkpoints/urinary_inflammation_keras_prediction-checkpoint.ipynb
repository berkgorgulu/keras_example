# Determining Target and Featues and Scaling

```python
y=df.iloc[:,-1]
X = df.iloc[:,:-1]
```

```python
X_numerical = pd.DataFrame(X.iloc[:,0])
X_categorical = pd.DataFrame(X.iloc[:,1:])
```

```python
# Getting dummies
y = pd.get_dummies(y, drop_first=True)

X_categorical = pd.get_dummies(X_categorical,drop_first=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Scale the data with `StandardScaler`

X_numerical_scaled =MinMaxScaler().fit_transform(X_numerical)

X.iloc[:,0] = X_numerical_scaled.reshape(-1,1)
```

```python
X.iloc[:,0] = X_numerical_scaled.reshape(-1,1)
```

```python
X = pd.concat([X_numerical,X_categorical],axis=1)
```

```python
# Import `train_test_split` from `sklearn.model_selection`
from sklearn.model_selection import train_test_split
# Split the data up in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Neural Network Architecture, Grid Search Optimizers

```python
from keras.optimizers import Adam
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import BatchNormalization
from sklearn.metrics import r2_score
from keras.regularizers import l2,l1
from keras.layers import Dropout,Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
from keras.models import Sequential
from keras import regularizers
import logging
logging.getLogger('tensorflow').disabled = True
```

```python
def make_keras_model(learning_rate=0.01,activation='relu',layer1_units=64, layer2_units=64,
                      dropout_rate=0.1, nbr_features=X_train.shape[1] ):
     
        # Describing optimization function
        opt = Adam(lr =learning_rate)  
        # Initialize the model
        model = Sequential()
        # Add input layer 
        model.add(Dense(layer1_units,input_shape=(nbr_features,),activation=activation))
        # Add first hidden layer
        model.add(Dense(layer2_units, activation = activation))
        # Add droupout layer
        model.add(Dropout(rate=dropout_rate))
        # Add output layer
        model.add(Dense(1,'sigmoid'))
    
        # Compile your model with your optimizer, loss, and metrics
        model.compile(optimizer = opt, loss = 'binary_crossentropy', metrics = ['accuracy'])
        return model
```

```python
classifiers_keras = KerasClassifier(make_keras_model, 
                                 batch_size=32, 
                                 layer1_units=64,
                                 layer2_units=64,
                                 dropout_rate=0,
                                 epochs=10, 
                                 verbose=False)
```

```python
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.impute import SimpleImputer
pipeline_keras = Pipeline([#('imputer', SimpleImputer()), 
                           #('scaler', StandardScaler()), 
                           ('model', classifiers_keras)])
```

```python
 param_grid_keras ={
    
    #'scaler': [StandardScaler()],
    'model__layer1_units': [32,128],
    'model__layer2_units': [32,64],
    'model__dropout_rate': [0, 0.2],
    'model__epochs': [10,75,150],
    'model__learning_rate':[0.01,0.1],
    'model__activation':['relu','tanh','softmax'],
    'model__batch_size':[32,64]
   
    
    
}
```

```python
from sklearn.model_selection import GridSearchCV

gridsearch_keras = GridSearchCV(pipeline_keras,
                                param_grid=param_grid_keras,
                                scoring='neg_log_loss',
                                cv=2, 
                                refit=True,
                                verbose=True)
```

```python
gridsearch_keras.fit(X_train, y_train)
```

```python
y_pred_keras = gridsearch_keras.predict(X_test)
```

```python
print(f'Best params: {gridsearch_keras.best_params_}')
```

```python
print(f'Best score: {gridsearch_keras.best_score_}')
```

```python
print(f'Test MSE: {mean_squared_error(y_true=y_test, y_pred=y_pred_keras)}')
```

```python
print(f'Test MAE: {mean_absolute_error(y_true=y_test, y_pred=y_pred_keras)}')
```

```python
r2_score = r2_score(y_test, y_pred_keras)
print('R2: ' '{:.4f}'.format(r2_score))
```

```python
gridsearch_keras.best_params_['model__batch_size']
```

```python
# Describing optimization function
opt = Adam(lr =gridsearch_keras.best_params_['model__learning_rate'])  
# Initialize the model
model_tuned = Sequential()
# Add input layer 
model_tuned.add(Dense(gridsearch_keras.best_params_['model__layer1_units'],input_shape=(X.shape[1],),activation=gridsearch_keras.best_params_['model__activation']))
# Add first hidden layer
model_tuned.add(Dense(gridsearch_keras.best_params_['model__layer2_units'], activation = gridsearch_keras.best_params_['model__activation']))
# Add droupout layer
model_tuned.add(Dropout(rate=gridsearch_keras.best_params_['model__dropout_rate']))
# Add output layer
model_tuned.add(Dense(1,'sigmoid'))
    
# Compile your model with your optimizer, loss, and metrics

model_tuned.compile(optimizer = opt, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model_tuned_hist = model_tuned.fit(X_train, y_train,epochs = gridsearch_keras.best_params_['model__epochs']
                                                 ,batch_size=gridsearch_keras.best_params_['model__batch_size'],
               validation_data=(X_test, y_test))
```

# Diagnostic and prediction

```python
# summarize history for accuracy
plt.plot(model_tuned_hist.history['accuracy'])
plt.plot(model_tuned_hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(model_tuned_hist.history['loss'])
plt.plot(model_tuned_hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
